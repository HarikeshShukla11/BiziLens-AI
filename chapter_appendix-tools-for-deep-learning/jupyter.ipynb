```python
import matplotlib.pyplot as plt
sales=[48000,54000,49000,57000,51000,45000,49500]
year=[2015,2016,2017,2018,2019,2020,2021]

plt.figure(figsize=(10,5))
plt.plot(year,sales)
plt.title("Samsung phones sales figure",color="g",fontsize=20,fontstyle="italic",fontweight="bold")
plt.xlabel("Years")
plt.ylabel("No of phones units sold")
plt.show()
```
