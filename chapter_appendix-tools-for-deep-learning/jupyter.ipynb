```python
import matplotlib.pyplot as plt

# Years from 2015 to 2025
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]

# Population in billions (approximate data + projections)
india_pop = [1.31, 1.33, 1.35, 1.36, 1.37, 1.38, 1.40, 1.41, 1.42, 1.43, 1.44]
china_pop = [1.37, 1.38, 1.39, 1.40, 1.40, 1.41, 1.41, 1.41, 1.41, 1.40, 1.39]

# Split data for actual (till 2021) and projected (20222025)
actual_years = years[:7]
future_years = years[7:]

india_actual = india_pop[:7]
india_future = india_pop[7:]

china_actual = china_pop[:7]
china_future = china_pop[7:]

# Figure setup
plt.figure(figsize=(9, 5))

# Actual data (solid line)
plt.plot(actual_years, india_actual, color='orange', linewidth=2, marker='o', label='India (Actual)')
plt.plot(actual_years, china_actual, color='red', linewidth=2, marker='s', label='China (Actual)')

# Future projections (dashed line)
plt.plot(future_years, india_future, color='orange', linewidth=2, linestyle='--', marker='o', label='India (Projected)')
plt.plot(future_years, china_future, color='red', linewidth=2, linestyle='--', marker='s', label='China (Projected)')

# Title and labels
plt.title("India vs China Population (20152025)",
          color="darkgreen", fontsize=18, fontstyle="italic", fontweight="bold")

plt.xlabel("Years", color='purple', fontsize=13, fontstyle="italic", fontweight="bold")
plt.ylabel("Population (in Billions)", color='purple', fontsize=13, fontstyle="italic", fontweight="bold")

# Legend and customizations
plt.legend(loc='upper left', fontsize=10)
plt.xticks(rotation=45, color='blue')
plt.yticks(color='blue')
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(1.28, 1.45)
plt.xlim(2015, 2025.5)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Years
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]

# Population data (in billions)  based on UN and Macrotrends estimates
india_pop = [1.31, 1.33, 1.35, 1.36, 1.37, 1.38, 1.40, 1.41, 1.43, 1.44, 1.45]
china_pop = [1.37, 1.38, 1.39, 1.40, 1.41, 1.41, 1.41, 1.41, 1.41, 1.42, 1.42]

# Create figure
plt.figure(figsize=(9, 5))

# Plot India
plt.plot(years, india_pop, color='orange', linewidth=2, linestyle='solid',
         marker='o', markersize=8, label='India')

# Plot China
plt.plot(years, china_pop, color='red', linewidth=2, linestyle='solid',
         marker='s', markersize=8, label='China')

# Titles and labels
plt.title("India vs China Population Comparison (20152025)",
          color="darkgreen", fontsize=18, fontstyle="italic", fontweight="bold")

plt.xlabel("Year", color='navy', fontsize=13, fontstyle="italic", fontweight="bold")
plt.ylabel("Population (in Billions)", color='navy', fontsize=13, fontstyle="italic", fontweight="bold")

# Add grid, legend, and axis limits
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='upper left', fontsize=11)
plt.xticks(rotation=45, color='blue')
plt.yticks(color='blue')
plt.ylim(1.28, 1.47)
plt.xlim(2015, 2025)

# Highlight year India surpasses China (2023)
plt.axvline(x=2023, color='green', linestyle='--', alpha=0.7)
plt.text(2023.1, 1.435, "India surpasses China ", color='green', fontsize=10, fontweight="bold")

# Show chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = np.arange(2015, 2026)
india_pop = [1.31, 1.33, 1.35, 1.36, 1.37, 1.38, 1.40, 1.41, 1.43, 1.44, 1.45]
china_pop = [1.37, 1.38, 1.39, 1.40, 1.41, 1.41, 1.41, 1.41, 1.41, 1.42, 1.42]

# Bar chart position setup
x = np.arange(len(years))
width = 0.35

plt.figure(figsize=(10, 6))

# Bar chart for populations
plt.bar(x - width/2, india_pop, width, color='orange', label='India')
plt.bar(x + width/2, china_pop, width, color='red', label='China')

# Line plots for trend
plt.plot(x, india_pop, color='brown', marker='o', linewidth=2)
plt.plot(x, china_pop, color='darkred', marker='s', linewidth=2)

# Title and labels
plt.title("India vs China Population Comparison (20152025)",
          fontsize=18, fontweight='bold', color='darkgreen')
plt.xlabel("Year", fontsize=13, fontweight='bold', color='navy')
plt.ylabel("Population (in Billions)", fontsize=13, fontweight='bold', color='navy')

# X-axis ticks
plt.xticks(x, years, rotation=45, color='blue')
plt.yticks(color='blue')

# Grid, legend
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
